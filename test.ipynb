```python
import json

with open('data/output.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 고유한 industry 추출
unique_industries = set()
for item in data:
    if 'industry' in item:
        unique_industries.add(item['industry'])

print(f"총 레코드 수: {len(data)}")
print(f"고유한 industry 개수: {len(unique_industries)}")
print("\n고유한 industry 목록:")
for industry in sorted(unique_industries):
    print(f"  - {industry}")
```
